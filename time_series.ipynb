```python
import pandas as pd
```

```python
df = pd.read_excel('TIME_SERIES.xlsx')
df
```

```python
#1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

state_to_plot = 'KERALA'
state_data = df[df['State'] == state_to_plot]

years = state_data['Year']
potable_count = state_data['Potable Count']
clean_count = state_data['Clean Count']

plt.figure(figsize=(10, 6))
plt.plot(years, potable_count, label='Potable Count', marker='o', color='g')
plt.plot(years, clean_count, label='Clean Count', marker='+', color='b')

plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(np.arange(min(df['Year']), max(df['Year'])+1,step=1))
plt.title(f'{state_to_plot} Potable and Clean Count Over the Years')
plt.legend()
plt.show()
```

```python
#2
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
states = df['State'].unique()
for state in states:
    state_data = df[df['State'] == state]
    years = state_data['Year']
    potable_count = state_data['Potable Count']
    plt.plot(years, potable_count, label=f'{state} - Potable Count', marker='o')

plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(np.arange(min(df['Year']), max(df['Year'])+1,step=1))
plt.title('Potable Count for All States Over the Years')
plt.legend(bbox_to_anchor=(1, 1.01))
plt.show()
```

```python
#4
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

south_data = df[df['Region'] == 'South']
south_states = south_data['State'].unique()

plt.figure(figsize=(12, 8))
for state in south_states:
    state_data = south_data[south_data['State'] == state]
    years = state_data['Year']
    potable_count = state_data['Potable Count']
    clean_count = state_data['Clean Count']

    # Plot "Potable Count" for each state in the North
    plt.plot(years, potable_count, label=f'{state} - Potable Count', marker='o')
    plt.plot(years, clean_count, label=f'{state} - Clean Count', marker='o')

# Add labels and a legend
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(np.arange(min(df['Year']), max(df['Year'])+1,step=1))
plt.title('Potable and Clean Count for States in the South Over the Years')
plt.legend(bbox_to_anchor=(1, 1.01))
plt.show()
```
