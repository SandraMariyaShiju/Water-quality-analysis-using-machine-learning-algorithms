```python
import pandas as pd
data = pd.read_excel('water_quality_2021corrected (1).xlsx')
print("The sample dataset is given below.")
data
```

```python
print("There are",len(data.axes[0]),"instances.")
#no:of variables
print("There are",len(data.axes[1]),"variables.")
#variable names
print('The columns of the dataset are \n', data.columns)
print('The information about the dataset is \n', data.info())
#describing the dataset
print('The description about the dataset is \n',data.describe())
```

```python
data1= data.drop(data.index[-1])
```

```python
data1
```

```python
data1.rename(columns={'STN Code':'Station_code'}, inplace=True)
data1.rename(columns={'Name of Monitoring Location':'Station_Name'}, inplace=True)
data1.rename(columns={'State Name':'State_Name'}, inplace=True)
data1.rename(columns={'Temperature Min':'Temp_Min'}, inplace=True)
data1.rename(columns={'Temperature Max ':'Temp_Max'}, inplace=True)
data1.rename(columns={'Dissolved Oxygen (mg/L)  Min':'Dissolved_O2_Min'}, inplace=True)
data1.rename(columns={'Dissolved Oxygen (mg/L) Max':'Dissolved_O2_Max'}, inplace=True)
data1.rename(columns={'pH Min':'pH_Min'}, inplace=True)
data1.rename(columns={'pH Max':'pH_Max'}, inplace=True)
data1.rename(columns={'Conductivity (μmho/cm) Min':'Conductivity_Min'}, inplace=True)
data1.rename(columns={'Conductivity (μmho/cm) Max':'Conductivity_Max'}, inplace=True)
data1.rename(columns={'BOD(mg/L) Min':'BOD_Min'}, inplace=True)
data1.rename(columns={'BOD(mg/L) Max':'BOD_Max'}, inplace=True)
data1.rename(columns={'Nitrate N + Nitrite N(mg/L)  Min':'Nitrate_Min'}, inplace=True)
data1.rename(columns={'Nitrate N + Nitrite N(mg/L)  Max':'Nitrate_Max'}, inplace=True)
data1.rename(columns={'Fecal Coliform (MPN/100ml) Min':'Fecal_Coliform_Min'}, inplace=True)
data1.rename(columns={'Fecal Coliform (MPN/100ml) Max':'Fecal_Coliform_Max'}, inplace=True)
data1.rename(columns={'Total Coliform (MPN/100ml) Min':'Total_Coliform_Min'}, inplace=True)
data1.rename(columns={'Total Coliform (MPN/100ml) Max':'Total_Coliform_Max'}, inplace=True)
```

```python
data1
```

```python
#checking for null values
data1.isnull().any()
```

```python
#replace BDL by 0
df=data1.replace(to_replace='BDL',value=0)
```

```python
df=data1.replace(to_replace='-',value=0)
df
```

```python
df.isnull().any()
```

```python
df=df.dropna()
df.isnull().any()
```

```python
#dataset after preprocessing
df
```
