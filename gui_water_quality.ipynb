```python
import pandas as pd
data = pd.read_excel('water_quality_2017corrected.xlsx')
print("The dataset is given below:")
data.head(5)
```

```python
#replace BDL by 0
df=data.replace(to_replace="BDL",value=0)
```

```python
df1=df.dropna()
df1.isnull().any()
```

```python
#Parameter threshold values
#Threshold values were collected from Research papers and websites
temp_min_threshold = 10.0
temp_max_threshold = 22.0
dissolved_O2_min_threshold = 6.5
dissolved_O2_max_threshold = 8.0
ph_min_threshold = 6.5
ph_max_threshold = 8.5
conductivity_min_threshold = 50
conductivity_max_threshold = 1500
bod_max_threshold = 3
nitrate_max_threshold = 45
total_coliform_max_threshold = 50
```

```python
# Create the Potability column based on rules
# Initialize as "Not Potable"
df1['Potability'] = 'Not Potable'
```

```python
df1.loc[(df1['Temp_Min'] < temp_min_threshold) |
 (df1['Temp_Max'] < temp_max_threshold) &
 (df1['Dissolved_O2_Max'] < dissolved_O2_max_threshold) &
 (df1['pH_Min'] < ph_min_threshold) |
 (df1['pH_Max'] < ph_max_threshold) &
 (df1['Conductivity_Min'] < conductivity_min_threshold) |
 (df1['Conductivity_Max'] < conductivity_max_threshold) &
 (df1['BOD_Max'] < bod_max_threshold) &
 (df1['Nitrate_Max'] < nitrate_max_threshold) &
 (df1['Total_Coliform_Max'] < total_coliform_max_threshold), 'Potability'] = 'Potable'
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['potab_label'] = le.fit_transform(df1['Potability'])
```

```python
df1.head(5)
```

```python
import tkinter as tk
from tkinter import *
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import pandas as pd

# Assume data is defined and contains the necessary information

# Extract Features and Target
X = df1.drop(["Station_code","Station_Name","State_Name","Potability","potab_label"], axis ='columns')
y = df1[['potab_label']]

# Split the data into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.30)

# Train an AdaBoostClassifier model
model=DecisionTreeClassifier()
model.fit(X_train, y_train)
#model.predict(X_test)

# Create a tkinter window
def my_reset():  # Define a function to reset the entry fields
    entry_temp_min.delete(0, tk.END)
    entry_temp_max.delete(0, tk.END)
    entry_dissolved_min.delete(0, tk.END)
    entry_dissolved_max.delete(0, tk.END)
    entry_ph_min.delete(0, tk.END)
    entry_ph_max.delete(0, tk.END)
    entry_conductivity_min.delete(0, tk.END)
    entry_conductivity_max.delete(0, tk.END)
    entry_bod_min.delete(0, tk.END)
    entry_bod_max.delete(0, tk.END)
    entry_nitrate_min.delete(0, tk.END)
    entry_nitrate_max.delete(0, tk.END)
    entry_f_coliform_min.delete(0, tk.END)
    entry_f_coliform_max.delete(0, tk.END)
    entry_t_coliform_min.delete(0, tk.END)
    entry_t_coliform_max.delete(0, tk.END)
    result_label.config(text="")


def on_exit():
    root.destroy()

root = tk.Tk()
root.title("Water Quality Prediction")

window_width = 1700  # Width in pixels
window_height = 900  # Height in pixels

root.geometry(f"{window_width}x{window_height}")  # Set the window size

#Frames
Tops = Frame(root, width=200, height=40, bd=12, relief="raise")
Tops.pack(side=TOP)
lblTitle = Label(Tops, font=("arial", 20, 'bold'), text="             Water Quality Prediction          ")
lblTitle.grid(row=0, column=0)

BottomMainFrame = Frame(root, width=600, height=90, bd=12, relief=SUNKEN)
BottomMainFrame.place(x=410,y=100)

f1left = Frame(BottomMainFrame, width=450, height=550, bd=4, relief='raise')
f1left.pack(side=LEFT)

f1middle= Frame(BottomMainFrame,width=450,height=550,bd=4,relief='raise')
f1middle.pack(side=LEFT)

f1right=Frame(BottomMainFrame,width=650,height=550,bd=4,relief='raise')
f1right.pack(side=LEFT)

# GUI input fields
label_temp_min = tk.Label(f1left, text="Temperature Minimum:")
label_temp_min.pack()
entry_temp_min = tk.Entry(f1left)
entry_temp_min.pack()


label_temp_max = tk.Label(f1left, text="Temperature Maximum:")
label_temp_max.pack()
entry_temp_max = tk.Entry(f1left)
entry_temp_max.pack()

label_dissolved_min = tk.Label(f1left, text="Dissolved O2 Minimum:")
label_dissolved_min.pack()
entry_dissolved_min = tk.Entry(f1left)
entry_dissolved_min.pack()

label_dissolved_max = tk.Label(f1left, text="Dissolved O2 Maximum:")
label_dissolved_max.pack()
entry_dissolved_max = tk.Entry(f1left)
entry_dissolved_max.pack()

label_ph_min = tk.Label(f1left, text="pH Minimum:")
label_ph_min.pack()
entry_ph_min = tk.Entry(f1left)
entry_ph_min.pack()

label_ph_max = tk.Label(f1left, text="pH Maximum:")
label_ph_max.pack()
entry_ph_max = tk.Entry(f1left)
entry_ph_max.pack()

label_conductivity_min = tk.Label(f1left, text="Conductivity Minimum:")
label_conductivity_min.pack()
entry_conductivity_min = tk.Entry(f1left)
entry_conductivity_min.pack()

label_conductivity_max = tk.Label(f1left, text="Conductivity Maximum:")
label_conductivity_max.pack()
entry_conductivity_max = tk.Entry(f1left)
entry_conductivity_max.pack()

label_bod_min = tk.Label(f1middle, text="BOD Minimum:")
label_bod_min.pack()
entry_bod_min = tk.Entry(f1middle)
entry_bod_min.pack()

label_bod_max = tk.Label(f1middle, text="BOD Maximum:")
label_bod_max.pack()
entry_bod_max = tk.Entry(f1middle)
entry_bod_max.pack()

label_nitrate_min = tk.Label(f1middle, text="Nitrate Minimum:")
label_nitrate_min.pack()
entry_nitrate_min = tk.Entry(f1middle)
entry_nitrate_min.pack()

label_nitrate_max = tk.Label(f1middle, text="Nitrate Maximum:")
label_nitrate_max.pack()
entry_nitrate_max = tk.Entry(f1middle)
entry_nitrate_max.pack()

label_f_coliform_min = tk.Label(f1middle, text="Fecal Coliform Minimum:")
label_f_coliform_min.pack()
entry_f_coliform_min = tk.Entry(f1middle)
entry_f_coliform_min.pack()

label_f_coliform_max = tk.Label(f1middle, text="Fecal Coliform Maximum:")
label_f_coliform_max.pack()
entry_f_coliform_max = tk.Entry(f1middle)
entry_f_coliform_max.pack()

label_t_coliform_min = tk.Label(f1middle, text="Total Coliform Minimum:")
label_t_coliform_min.pack()
entry_t_coliform_min = tk.Entry(f1middle)
entry_t_coliform_min.pack()

label_t_coliform_max = tk.Label(f1middle, text="Total Coliform Maximum:")
label_t_coliform_max.pack()
entry_t_coliform_max = tk.Entry(f1middle)
entry_t_coliform_max.pack()

b1 = tk.Button(f1right, text='Reset', font=5, bg='white', command=my_reset, width=20, height=1)
b1.pack()

def predict_event():
    temp_min = float(entry_temp_min.get())
    temp_max = float(entry_temp_max.get())
    dissolved_min = float(entry_dissolved_min.get())
    dissolved_max = float(entry_dissolved_max.get())
    ph_min = float(entry_ph_min.get())
    ph_max = float(entry_ph_max.get())
    conductivity_min = float(entry_conductivity_min.get())
    conductivity_max = float(entry_conductivity_max.get())
    bod_min = float(entry_bod_min.get())
    bod_max = float(entry_bod_max.get())
    nitrate_min = float(entry_nitrate_min.get())
    nitrate_max = float(entry_nitrate_max.get())
    f_coliform_min = float(entry_f_coliform_min.get())
    f_coliform_max = float(entry_f_coliform_max.get())
    t_coliform_min = float(entry_t_coliform_min.get())
    t_coliform_max = float(entry_t_coliform_max.get())

    event = model.predict([[temp_min, temp_max, dissolved_min, dissolved_max, ph_min, ph_max,
                           conductivity_min, conductivity_max, bod_min, bod_max, nitrate_min, nitrate_max,
                           f_coliform_min, f_coliform_max, t_coliform_min, t_coliform_max]])
    # Inverse transform the label encoding to get the original event type
    predicted_event = le.inverse_transform(event)

    result_label.config(text=f"The Predicted Water Quality is: {predicted_event[0]}")


predict_button = tk.Button(f1right, text="Predict", command=predict_event, font=4, bg='white', width=20, height=1)
predict_button.pack()
result_label = tk.Label(f1right, font=('bold'), text="", width=40, height=3)
result_label.pack()
exit_button = tk.Button(f1right, text="Exit", command=on_exit, font=5, bg='white', width=20, height=1)
exit_button.pack()

root.mainloop()
```
