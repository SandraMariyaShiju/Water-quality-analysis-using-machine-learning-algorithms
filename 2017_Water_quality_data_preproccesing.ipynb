```python
import pandas as pd
data = pd.read_excel('water_quality_2017corrected (1).xlsx')
print("The sample dataset is given below.")
data
```

```python
print("There are",len(data.axes[0]),"instances.")
#no:of variables
print("There are",len(data.axes[1]),"variables.")
#variable names
print('The columns of the dataset are \n', data.columns)
print('The information about the dataset is \n', data.info())
#describing the dataset
print('The description about the dataset is \n',data.describe())
```

```python
data.rename(columns={'Station code':'Station_code'}, inplace=True)
data.rename(columns={'Station Name':'Station_Name'}, inplace=True)
data.rename(columns={'State Name':'State_Name'}, inplace=True)
data.rename(columns={'Temperature Min':'Temp_Min'}, inplace=True)
data.rename(columns={'Temperature Max':'Temp_Max'}, inplace=True)
data.rename(columns={'DissolvedO2 mg/l Min':'Dissolved_O2_Min'}, inplace=True)
data.rename(columns={'DissolvedO2 mg/l Max':'Dissolved_O2_Max'}, inplace=True)
data.rename(columns={'pH Min':'pH_Min'}, inplace=True)
data.rename(columns={'pH Max':'pH_Max'}, inplace=True)
data.rename(columns={'Conductivity (µmhos/cm)  Min':'Conductivity_Min'}, inplace=True)
data.rename(columns={'Conductivity (µmhos/cm) Max':'Conductivity_Max'}, inplace=True)
data.rename(columns={'BOD mg/l Min':'BOD_Min'}, inplace=True)
data.rename(columns={'BOD mg/l Max':'BOD_Max'}, inplace=True)
data.rename(columns={'Nitrate N +Nitrite N mg/l Min ':'Nitrate_Min'}, inplace=True)
data.rename(columns={'Nitrate N  Nitrite N mg/l  Max':'Nitrate_Max'}, inplace=True)
data.rename(columns={'Fecal Coliform MPN/100ML   Min':'Fecal_Coliform_Min'}, inplace=True)
data.rename(columns={'Fecal Coliform MPN/100ML Max':'Fecal_Coliform_Max'}, inplace=True)
data.rename(columns={'Total Coliform MPN/100ML Min':'Total_Coliform_Min'}, inplace=True)
data.rename(columns={'Total Coliform MPN/100ML Max':'Total_Coliform_Max'}, inplace=True)
```

```python
#replace BDL by 0
df=data.replace(to_replace="BDL",value=0)
```
